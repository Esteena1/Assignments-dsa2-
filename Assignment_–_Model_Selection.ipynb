1. Load the dataset into python environment

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/titanic_dataset .csv")
df.head()
df.info()
```

2. Do all the necessary pre-processing steps

```python
df1=df.dropna()
df1.info()
df1.head()
```

3. Create kNN and SVM models

```python
feature_columns = ['SibSp','Parch',	'Fare']
X = df1[feature_columns].values
y = df1['Embarked'].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fitting clasifier to the Training set
from sklearn.neighbors import KNeighborsClassifier

# Instantiate learning model (k = 3)
classifier = KNeighborsClassifier(n_neighbors=3)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

# 4.Do k-fold and stratified stratified k-fold cross validation techniques and find the average accuracy score of the models

```python
df1=pd.get_dummies(df1)
df1.head()
```

```python
x=df1.drop('Survived',axis=1)
y=df1['Survived']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
from sklearn.linear_model import LogisticRegression
LR_model=LogisticRegression()
LR_model=LR_model.fit(x_train,y_train)
score_LR=LR_model.score(x_test,y_test)
score_LR

from sklearn.model_selection import KFold
kfold_validator=KFold(10)
for train_index,test_index in kfold_validator.split(x,y):
  print('Training Index:',train_index)
  print('Testing Index:',test_index)
from sklearn.model_selection import cross_val_score
cv_result=cross_val_score(LR_model,x,y,cv=kfold_validator)
cv_result
np.mean(cv_result)
```

```python
## Stratified Cross Validation
from sklearn.model_selection import StratifiedKFold
skfold_validator=StratifiedKFold(n_splits=2)
for train_index,test_index in skfold_validator.split(x,y):
  print('Training Index:',train_index)
  print('Testing index:',test_index)
cv_result=cross_val_score(LR_model,x,y,cv=kfold_validator)
cv_result
np.mean(cv_result)
```
