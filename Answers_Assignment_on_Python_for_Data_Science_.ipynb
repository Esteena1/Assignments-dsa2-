```python
import matplotlib.pyplot as plt
import seaborn as sns

### Scatterplot between Sepal Length and Sepal Width
data.plot(kind="scatter",x='SL',y='SW')
```

```python
### Jointplot between PetalLength and PetalWidth
sns.jointplot(x="PL",y="PW",data=data)
```

```python
### Boxplot
sns.boxplot(x='Classification',y='PW', data=data)
```

```python
### Violinplot
sns.violinplot(x='Classification',y='PL',data=data)
```

```python
### pairplot

sns.pairplot(data=data)
```
